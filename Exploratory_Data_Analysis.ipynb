```python
# import the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# load dataset 
athletes = pd.read_csv('athlete_events.csv')
region = pd.read_csv('noc_regions.csv')
```

```python
athletes.head()
```

```python
region.head()
```

```python
# join the dataframes
athletes_df = athletes.merge(region, how='left', on='NOC')
athletes_df.head()
```

```python
athletes_df.shape
```

```python
athletes_df.rename(columns={'region':'Region','notes':'Notes'}, inplace=True)
athletes_df.head()
```

```python
# detail information about the dataframe
athletes_df.info()
```

```python
athletes_df.describe()
```

```python
# check null values
nan_values = athletes_df.isna()
nan_columns = nan_values.any()
nan_columns
```

```python
# display null values in total
athletes_df.isnull().sum()
```

```python
# print the columns names containing null values in the form of a list
athletes_list = athletes_df.columns[athletes_df.isnull().any()].tolist()
athletes_list
```

```python
# india details
athletes_df.query('Team == "India"').head(2)
```

```python
# top countries participating
top_10_countries = athletes_df.Team.value_counts().sort_values(ascending=False).head(10)
top_10_countries
```

```python
# plot for the top 10 countries
plt.figure(figsize=(12,6))
plt.title('Overall Participation by country')
sns.barplot(x=top_10_countries.index, y=top_10_countries, palette='Set2')
```

```python
# Age Distribution of the participants
plt.figure(figsize=(12, 6))
plt.title("Age distribution of the athletes")
plt.xlabel('Age')
plt.ylabel('Number of participants')
plt.hist(athletes_df.Age, bins=np.arange(10,80,2), color='green', edgecolor='white')
```

```python
# details for winter as season sports details
winter_sports=athletes_df[athletes_df.Season=='Winter'].Sport.unique()
winter_sports.tolist()
```

```python
# details for summer as season sports details
summer_sports=athletes_df[athletes_df.Season=='Summer'].Sport.unique()
summer_sports.tolist()
```

```python
# male and female participants
gender_counts = athletes_df.Sex.value_counts()
gender_counts
```

```python
# pie plot for male and female athletes
plt.figure(figsize=(12,6))
plt.title('Gender Distribution')
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.2f%%', startangle=150, shadow=True)
```

```python
# total medals
athletes_df.Medal.value_counts()
```

```python
# total number of female athletes in each olympics
female_participants = athletes_df[(athletes_df.Sex=='F') & (athletes_df.Season=='Summer')][['Sex','Year']]
female_participants = female_participants.groupby('Year').count().reset_index()
female_participants.tail()
```

```python
womenOlympics = athletes_df[(athletes_df.Sex=='F') & (athletes_df.Season=='Summer')]
```

```python
# plot for womenOlympics in summer season
sns.set(style="darkgrid")
plt.figure(figsize=(12, 8))
sns.countplot(x='Year', data=womenOlympics, palette="Spectral")
plt.title('Women Participation')
```

```python
# plot for counts of female participants increase over time
part = womenOlympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(12, 6))
part.loc[:,'F'].plot()
plt.title('Plot of female athletes over time')
```

```python
# gold medal athletes
goldMedals = athletes_df[(athletes_df.Medal=='Gold')]
goldMedals.head()
```

```python
# take only the values that are different from NaN
goldMedals = goldMedals[np.isfinite(goldMedals['Age'])]
goldMedals.head()
```

```python
# Gold beyond 60
goldMedals['ID'][goldMedals['Age'] > 60].count()
```

```python
# list of sports as age > 60
sporting_event = goldMedals['Sport'][goldMedals['Age']>60]
sporting_event
```

```python
# plot for sporting_event
plt.figure(figsize=(12, 6))
plt.tight_layout()
sns.countplot(sporting_event)
plt.title('Gold Medals for athletes over 60 years')
```

```python
# gold medals from each country
goldMedals.Region.value_counts().reset_index(name='Medal').head()
```

```python
# Gold Medals per country
totalGoldMedals = goldMedals.Region.value_counts().reset_index(name='Medal').head()
g = sns.catplot(x="index", y="Medal", data=totalGoldMedals, height=5, kind="bar", palette="rocket")
g.despine(left=True)
g.set_xlabels("Total 5 countries")
g.set_ylabels("Number of Medals")
plt.title("Gold Medals per country")
```

```python
# Rio olympics
max_year=athletes_df.Year.max()
print(max_year)
team_names=athletes_df[(athletes_df.Year==max_year) & (athletes_df.Medal=='Gold')].Team
team_names.value_counts().head()
```

```python
# Countrywise Medals for the year 2016
sns.barplot(x=team_names.value_counts().head(15), y=team_names.value_counts().head(15).index)
plt.ylabel(None)
plt.xlabel('Countrywise Medals for the year 2016')
```

```python
not_null_medals=athletes_df[(athletes_df['Height'].notnull()) & (athletes_df['Weight'].notnull())]
```

```python
# Height vs Weight of Olympics Medalists
plt.figure(figsize=(12, 10))
axis=sns.scatterplot(x="Height",y="Weight", data=not_null_medals, hue="Sex")
plt.title('Height vs Weight of Olympics Medalists')
```
